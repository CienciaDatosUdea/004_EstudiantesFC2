```python
import numpy as np
import matplotlib.pyplot as plt
from celluloid import Camera
from IPython.display import HTML
```

```python
def animate(file,Nx,save):   
    # Lectura y ordenamiento de los datos 
    data = np.loadtxt(file,comments='#')
    xs = data[:,0]
    Ts = data[:,1]
    Ny = xs.size//Nx
    xs = xs.reshape((Ny,Nx))
    Ts = Ts.reshape((Ny,Nx))

    # Creación de la gráfica
    fig,ax = plt.subplots()
    cam = Camera(fig)
    for i in range(Ny):
        ax.plot(xs[i],Ts[i],'r')
        ax.imshow(
            Ts[i][np.newaxis, :],  # Banda de 1 pixel de alto
            aspect='auto',
            cmap='inferno',
            vmin=0,vmax = 1,
            extent=[xs[i].min(), xs[i].max(), ax.get_ylim()[0], ax.get_ylim()[0] + 0.05*(ax.get_ylim()[1]-ax.get_ylim()[0])]
        )
        cam.snap()
    
    ax.set(title='Propagación del calor en una barra 1D',
           xlabel='x',
           ylabel='T')
    ax.grid()
    ax.axis('equal')
    plt.close()

    anim = cam.animate(interval = 40)
    if save:
        anim.save('anim.gif')
    return HTML(anim.to_html5_video())
```

```python
# Poner la ruta donde se encuentra el archivo que se va a animar

animate('/home/santiago/santiago_FCII_20251/Tarea_02/data/solucion_final.txt',101,1)
```
