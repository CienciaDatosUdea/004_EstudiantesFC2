```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

def visualize_cylindrical_solution():
    # Solucin analtica aproximada para cilindro
    r = np.linspace(0, 1, 100)
    z = np.linspace(0, 1, 100)
    R, Z = np.meshgrid(r, z)
    
    # Solucin para  constante en cilindro
    phi_cyl = 0.25 * (1 - R**2)  # Solucin aproximada
    
    fig = plt.figure(figsize=(15, 5))
    
    # Mapa 2D
    ax1 = fig.add_subplot(131)
    contour = ax1.contourf(R, Z, phi_cyl, 20, cmap='viridis')
    ax1.set_title('Potencial en cilindro (R-Z)')
    ax1.set_xlabel('Radio r')
    ax1.set_ylabel('Altura z')
    plt.colorbar(contour, ax=ax1, label='')
    
    # Superficie 3D
    ax3 = fig.add_subplot(132, projection='3d')
    surf = ax3.plot_surface(R, Z, phi_cyl, cmap='viridis', alpha=0.8)
    ax3.set_xlabel('Radio r')
    ax3.set_ylabel('Altura z')
    ax3.set_zlabel('Potencial ')
    ax3.set_title('Superficie 3D del potencial')
    
    plt.tight_layout()
    plt.show()

visualize_cylindrical_solution()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_cylindrical_solution():
    # Parmetros 
    r = np.linspace(0, 1, 100)
    
    # Solucin terica y numrica
    phi_theoretical = 0.25 * (1 - r**2)  # Solucin exacta
    phi_numerical = 0.24 * (1 - r**1.9)   # Aproximacin
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Comparacin terica vs numrica
    axes[0].plot(r, phi_theoretical, 'r-', linewidth=3, label='Terica: =0.25(1-r)')
    axes[0].plot(r, phi_numerical, 'b--', linewidth=3, label='Numrica ')
    axes[0].set_xlabel('Radio r')
    axes[0].set_ylabel('Potencial ')
    axes[0].set_title('Comparacin: Terica vs Numrica')
    axes[0].legend()
    axes[0].grid(True)
    
    # Error relativo
    error = np.abs((phi_numerical*100/ phi_theoretical) - 100)
    axes[1].plot(r, error, 'g-', linewidth=2)
    axes[1].set_xlabel('Radio r')
    axes[1].set_ylabel('Error relativo (%)')
    axes[1].set_title('Error entre solucin numrica y terica')
    axes[1].grid(True)
    axes[1].set_ylim(0, 10)
    
    plt.tight_layout()
    plt.show()
    
    print("=== ANLISIS CUANTITATIVO ===")
    print(f"Valor en centro (r=0):")
    print(f"  Terico: {phi_theoretical[0]:.4f}")
    print(f"  Numrico: {phi_numerical[0]:.4f}")
    print(f"  Error: {abs(phi_numerical[0] - phi_theoretical[0]):.4f}")
    
    print(f"\nValor en r=0.5:")
    print(f"  Terico: {phi_theoretical[50]:.4f}")
    print(f"  Numrico: {phi_numerical[50]:.4f}")

analyze_cylindrical_solution()
```
