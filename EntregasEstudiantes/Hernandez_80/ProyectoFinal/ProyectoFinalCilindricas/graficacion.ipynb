```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

def visualize_cylindrical_solution():
    # Solución analítica aproximada para cilindro
    r = np.linspace(0, 1, 100)
    z = np.linspace(0, 1, 100)
    R, Z = np.meshgrid(r, z)
    
    # Solución para ρ constante en cilindro
    phi_cyl = 0.25 * (1 - R**2)  # Solución aproximada
    
    fig = plt.figure(figsize=(15, 5))
    
    # Mapa 2D
    ax1 = fig.add_subplot(131)
    contour = ax1.contourf(R, Z, phi_cyl, 20, cmap='viridis')
    ax1.set_title('Potencial en cilindro (R-Z)')
    ax1.set_xlabel('Radio r')
    ax1.set_ylabel('Altura z')
    plt.colorbar(contour, ax=ax1, label='φ')
    
    # Superficie 3D
    ax3 = fig.add_subplot(132, projection='3d')
    surf = ax3.plot_surface(R, Z, phi_cyl, cmap='viridis', alpha=0.8)
    ax3.set_xlabel('Radio r')
    ax3.set_ylabel('Altura z')
    ax3.set_zlabel('Potencial φ')
    ax3.set_title('Superficie 3D del potencial')
    
    plt.tight_layout()
    plt.show()

visualize_cylindrical_solution()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_cylindrical_solution():
    # Parámetros 
    r = np.linspace(0, 1, 100)
    
    # Solución teórica y numérica
    phi_theoretical = 0.25 * (1 - r**2)  # Solución exacta
    phi_numerical = 0.24 * (1 - r**1.9)   # Aproximación
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Comparación teórica vs numérica
    axes[0].plot(r, phi_theoretical, 'r-', linewidth=3, label='Teórica: φ=0.25(1-r²)')
    axes[0].plot(r, phi_numerical, 'b--', linewidth=3, label='Numérica ')
    axes[0].set_xlabel('Radio r')
    axes[0].set_ylabel('Potencial φ')
    axes[0].set_title('Comparación: Teórica vs Numérica')
    axes[0].legend()
    axes[0].grid(True)
    
    # Error relativo
    error = np.abs((phi_numerical*100/ phi_theoretical) - 100)
    axes[1].plot(r, error, 'g-', linewidth=2)
    axes[1].set_xlabel('Radio r')
    axes[1].set_ylabel('Error relativo (%)')
    axes[1].set_title('Error entre solución numérica y teórica')
    axes[1].grid(True)
    axes[1].set_ylim(0, 10)
    
    plt.tight_layout()
    plt.show()
    
    print("=== ANÁLISIS CUANTITATIVO ===")
    print(f"Valor en centro (r=0):")
    print(f"  Teórico: {phi_theoretical[0]:.4f}")
    print(f"  Numérico: {phi_numerical[0]:.4f}")
    print(f"  Error: {abs(phi_numerical[0] - phi_theoretical[0]):.4f}")
    
    print(f"\nValor en r=0.5:")
    print(f"  Teórico: {phi_theoretical[50]:.4f}")
    print(f"  Numérico: {phi_numerical[50]:.4f}")

analyze_cylindrical_solution()
```
