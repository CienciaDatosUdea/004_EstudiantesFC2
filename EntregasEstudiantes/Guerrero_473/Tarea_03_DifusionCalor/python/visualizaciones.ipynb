```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import glob
import os
```

```python
#Evolución a partir de un "pulso" gaussiano como condición inicial

ruta_datos="datos"
nombre_archivo="gaussiana.txt"
archivo = os.path.join(ruta_datos, nombre_archivo)
data = np.loadtxt(archivo, delimiter=',', skiprows=1)

data = np.loadtxt(archivo, delimiter=',', skiprows=1)

tiempos = np.unique(data[:, 0])   # tiempos únicos
xs = np.unique(data[:, 1])        # posiciones únicas

# Reorganizar datos en matriz (len(xs) x len(tiempos))
U = np.zeros((len(xs), len(tiempos)))
for i, t in enumerate(tiempos):
    mask = data[:, 0] == t
    U[:, i] = data[mask, 2]

# Graficar
plt.figure(figsize=(8, 6))
plt.imshow(U, aspect='auto', origin='lower',
           extent=[tiempos.min(), tiempos.max(), xs.min(), xs.max()],
           cmap='hot')
plt.colorbar(label='u(x)')
plt.xlabel('Tiempo')
plt.ylabel('x')
plt.title('Evolución temporal de u(x)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for t in [0, 0.1,0.2,0.3 ,0.5, 1.0]:
    mask = np.isclose(data[:, 0], t)
    plt.plot(data[mask, 1], data[mask, 2], label=f't={t}')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Perfiles a diferentes tiempos')
plt.legend()
plt.grid()
plt.show()
```
